```python
ANS 1: Matplotlib is a popular data visualization library for the Python programming language. It provides a wide variety of 
tools for craeting static, interactive and animated visualization in Python. Matploylib is widely used in the data science, 
scientific research, and engineering communities to effectively communicate insights and findings from data.

Five types of plot that can be craeted using the 'Pyplot' module of Matplotlib:
1. Line plot : This is the most basic type of plot where data points are connected with lines. It's often used to display 
trends or patterns over time or other continuous variables.
2. Scatter plot: A scatter plot displays individual data points as dot on a two-dimensional plane. It's used to show the 
relationship between two continuous variables and helps identify patterns or cluster in a data.
3. Bar plot: Bar plots represent data with rectangular bars, where the length or height of each bar corresponds to the value of
the data it represents. They are commonly used to compare categorical data. 
4. Histogram : Histograms are used to visualize the distribution of a continuous variable. They divide the data into bins and
display the frequency or count of data points within each bin.
5.Pie Chart: A pie chart is a circular chart that represents data as slices of a pie. Each slice's size is proportional to the
data's value it represents. Pie charts are useful for displaying the composition of a whole in terms of its parts.
```

```python
ANS 2: A scatter plot is a type of data visualization that uses dots to represent the values obtained for two different
variables in a dataset. Each dot on the scatter plot corresponds to a single data point, with its position on the x-axis and 
y-axis determined by the values of the two variables being compared.
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3+np.random.normal(0,2,50)
y = 3+np.random.normal(0,2,len(x))

plt.scatter(x,y)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")

plt.show()
```

```python
ANS 3: The 'subplot()' function in Matplotlib is used to create multiple plots within a single figure. It allows you to arrange
multiple plots in a grid-like fashion, making it easier to compare and analyze different visualizations.
```

```python
import numpy as pd 
import matplotlib.pyplot as plt

x = np.array([0,1,2,3,4,5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("Line 1")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("Line 2")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("Line 3")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("Line 4")

plt.tight_layout()
plt.show()
```

```python
ANS 4: A bar plot is a type of data visualization that uses rectangular bars to represent data values. Each bar's length or 
height corresponds to the value of the data it represents. Bar plots are commonly used to compare different categories or
groups of data, showing the relative sizes of the values and highlighting any patterns or trends within the dataset.
```

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(6,4))
plt.bar(company,profit,color = 'blue')
plt.title("Company Profits")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

plt.figure(figsize=(7,5))
plt.barh(company,profit,color='green')
plt.title("Company Profits (Horizontal)")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

```python
ANS 5: A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset's 
values. It provides a summary of the data's central tendency, dispersion, and any potential outliers. The box in the plot
represents the interquartile range (IQR), which contains the middle 50% of the data. The whiskers extend from the box to the
minimum and maximum values within a certain range (often 1.5 times the IQR), and any data points beyond the whiskers are
considered outliers.
Box plots are used for several purposes:
1.Identifying Distribution: Box plots provide a clear visualization of the data's distribution, including the median,
quartiles, and potential outliers. This information helps in understanding the overall pattern and variability of the data.
2.Comparing Distributions: Box plots can be used to compare the distributions of multiple datasets, making it easier to spot
differences in central tendency, spread, and presence of outliers.
3.Detecting Outliers: Outliers, which are values significantly different from the rest of the data, are easy to identify in a 
box plot. This can be useful for data cleaning and anomaly detection.
4.Highlighting Variability: Box plots display the variability of the data, indicating whether the values are tightly clustered
or widely spread.
5.Summarizing Data: Box plots provide a concise summary of key statistics, making them suitable for presentations and quick 
comparisons.
```

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)
data = [box1,box2]

plt.figure(figsize=(8,5))
plt.boxplot(data , labels=["Box1","Box2"])
plt.title("Box plot of two databases")
plt.ylabel("Values")
plt.show()
```

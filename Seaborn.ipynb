```python
ANS 1: five plots that we can plot using the Seaborn library and their uses:
1.Scatter Plot:Scatter plots are used to display the relationship between two continuous variables. Each point on the 
plot represents a data point with its x and y values.
2. Line Plot: Line plots are suitable for showing trends over continuous data points, such as time series data. They can
be used to visualize the change in a variable's value over a specific interval.
3. Bar Plot: Bar plots are used to compare categorical data or display the distribution of a single categorical variable.
They can show the frequency, count, or average of different categories.
4. Histogram: Histograms are used to visualize the distribution of a single numeric variable. They show the frequency of
values falling into predefined bins or intervals, which helps to understand the underlying data distribution.
5. Box Plot: Box plots (box-and-whisker plots) are used to visualize the distribution of a dataset's summary statistics,
such as median, quartiles, and potential outliers. They are particularly useful for comparing distributions across
different groups or categories.
```

```python
import seaborn as sns

fmri = sns.load_dataset("fmri")
sns.lineplot(x = fmri.timepoint , y = fmri.signal , hue = fmri.event , style = fmri.region ,  markers=True , data = fmri)
```

```python
import seaborn as sns
pclass_order = [1,2,3]

titanic = sns.load_dataset("titanic")
g = sns.FacetGrid(titanic, col="pclass", height=6, aspect=0.8)
g.map(sns.boxplot,"pclass","age",order=pclass_order)
g.set_titles("Box plot of age by pclass")

g = sns.FacetGrid(titanic,col="pclass",height=6,aspect=0.8)
g.map(sns.boxplot,"pclass","fare",order=pclass_order)
g.set_titles("Box plot of fare by pclass")
```

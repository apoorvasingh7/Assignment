```python
ANS 1: GET method: 
    The GET method is used to retrieve or fetch data from a server. When a client makes a GET request, it includes parameters
    in the URL of the request, usually after a question mark('?'). These parameters are in the form of key-value pairs. The 
    server processes the request and returns the requested data in the response body.
POST Method:
    The POST method is used to submit or send data to a server to create or update the resource. Unlike the GET method, the
    parameters and data in a POST request are sent in the body of the request rather than the URL. The data can be in various
    formats such as JSON, XML, or form data.
```

```python
ANS 2: In the Flask web framework, the 'request' object is used to access the incoming client request data. It provides a way
to retrieve information from the client, such as form data, query parameters, headers, cookies and more.
The 'request' object encapsulates the details of the HTTP request made by the client and provides methods and attributes to 
access and manipulate the request data.
```

```python
ANS 3: The 'rediret()' function in Flask is used to perform a redirect operation, sending the client's browser to a different 
URL. It is convenient way to redirect the user to another route or an external URL.
```

```python
ANS 4: In Flask, templates are files that contain the structure and layout of the web pages or email messages to be rendered 
dynamically. Templates allow you to separate the presentation logicfrom the application logic, making it easier to maintain and
modify the user interface of your web application.
The 'render_template()' function in Flask is used to render a template and generate the final HTML output to be sent as a 
response to the client's browser. It takes the name of the template files as an argument and optionally accepts additional 
paremeters to pass data to the template.
```

```python
ANS 5: from flask import Flask, jsonify, request

app = Flask(__name__)

@app.route("/api/hello", methods=['GET'])
def hello():
    return jsonify(message='Hello World!')

@app.route("/api/greeting", methods=['POST'])
def greeting():
    name = request.json.get('name')
    if name:
        return jsonify(message=f"Hello, {name}!")
    else:
        return jsonify(error='Name parameter is missing.')

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

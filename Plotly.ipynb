```python
import plotly.graph_objects as go
import seaborn as sns
titanic = sns.load_dataset('titanic')
fig = go.Figure()
fig.add_trace(go.Scatter(x=titanic.age , y=titanic.fare , mode='markers'))
fig.show()
```

```python
import seaborn as sns
import plotly.express as px
tips_df = sns.load_dataset("tips")
box_plot = px.box(tips_df, x='day', y='total_bill', title='Box Plot of Total Bill Amounts by Day')
box_plot.show()
```

```python
import seaborn as sns
import plotly.express as px
tips = sns.load_dataset('tips')
histogram = px.histogram(tips,x=tips.sex,y=tips.total_bill,pattern_shape=tips.smoker,color=tips.day)
histogram.show() 
```

```python
import seaborn as sns 
import plotly.express as px
iris = sns.load_dataset('iris')
matrix = px.scatter_matrix(iris,dimensions = ['sepal_length','sepal_width','petal_length','petal_width'],color='species')
matrix.show() 
```

```python
ANS 5: The 'displot' function (distribution plot) is a part of the Seaborn library, used for creating visualizations that
represent the distribution of a dataset. It combines elements from histograms and kernel density estimation (KDE) plots to 
provide insights into the underlying data distribution. In a displot, data is usually represented as bars (in the case of a
histogram) or as a smooth curve (in the case of a KDE plot) to visualize how the data is spread across different values.
The 'displot' function has been deprecated in Seaborn starting from version 0.11.0. The recommended alternatives are to use 
'histplot' for histogram-based visualizations and 'kdeplot' for kernel density estimation-based visualizations.
```

```python
import seaborn as sns
import plotly.express as px
tips = sns.load_dataset('tips')
his_plot = px.histogram(tips,x=tips.total_bill)
his_plot.show()
```

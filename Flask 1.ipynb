```python
ANS 2: 
    
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "</h1>Hello World!!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
ANS 3: In Flask, app routing refers to the process of mapping URL paths to specific functionsin your application. It 
allows you to define different routes(URL paths) and associate them with corresponding functions that handle the logic 
and generate the response  for that partiular route.
the '@app.route()' decorator in Flask is used to define routes. It is a way to bind a URL pattern to a function,
indicating which function should be executed when a specific URL is requested.
```

```python


ANS 4:
    
from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "Welcome to ABC Corporation"

@app.route("/")
def company_details():
    company_name = "ABC Corporation"
    location = "India"
    contact_details = 9999999999

    details=f"Company Name: {company_name}<br>"
    details+=f"Location: {location}<br>"
    details+=f"Contact Details: {contact_details}"

    return details

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
ANS 5: In Flask, the 'url_for()' function is used for URL building. 
  
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return "This is the home page"

@app.route('/user/<username>')
def profile(username):
    return f"Welcome, {username}!"

@app.route('/product/<int:product_id>')
def product(product_id):
    return f"Product ID: {product_id}"

@app.route('/login')
def login():
    home_url = url_for('home')
    profile_url = url_for('profile',username='john')
    product_url = url_for('product',product_id=123)
    
    return f"Home URL: {home_url}<br>"\
           f"Profile URL: {profile_url}<br>"\
           f"Product URL: {product_url}"

if __name__=='__main__':
    app.run()
    
    
In this example, the 'url_for()' function is used inside the 'login()' function to generate URLs for different endpoints.
The 'home_url' variable generates the URL for the 'home' endpoint, the 'profile_url' variable generates the URL for the 
'profile' endpoint with the 'john' username parameter, and the 'product_url' variable generate the URL for the 'product' 
endpoint with the '123' product_id parameter.
```
